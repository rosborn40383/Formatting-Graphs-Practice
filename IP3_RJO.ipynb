```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import statsmodels.api as sm
```

Defining a function to read the csv's about wind turbine data, then add a column fro the ocmpany and convert to a date/time

```python
def convert(filename):
    df = pd.read_csv(filename)
    df.Date = pd.to_datetime(df.Date)
    df.insert(loc = 0, column = 'Company', value = (filename[:-4]))
    return(df)
```

```python
hertz=convert('50Hertz.csv')
amp=convert('Amprion.csv')
tennet=convert('TenneTTSO.csv')
trans=convert('TransnetBW.csv')
```

Concatenate the data frames

```python
conc_t=pd.concat([hertz,amp,tennet,trans])
```

```python
conc_t
```

```python
conc_mean=conc_t.groupby([conc_t.Company, conc_t.Date.dt.to_period("Y")]).mean()
```

```python
conc_mean
```

Creating a bar (horizontal) plot of the data

```python
midnight=conc_mean['00:00:00']
noon=conc_mean['12:00:00']
```

```python
midnight_barh=plt.figure()
midnight.sort_values().plot(kind='barh')
plt.title('TW/hour by Company/Year @ midnight')
```

```python
noon_barh=plt.figure()
noon.sort_values().plot(kind='barh', color='red')
plt.title('TW/hour by Company/Year @ noon')
```

Putting two bar graphs into single bar graph

```python
combined_barh, (ax_mid, ax_noon) = plt.subplots(2, 1, sharex=True)
midnight.sort_values().plot(kind='barh', ax=ax_mid, color='silver', title='TW/hour by Company/Year @ midnight')
noon.sort_values().plot(kind='barh', ax=ax_noon, color='gray', title='TW/hour by Company/Year @ noon')
plt.savefig('combined_bar_chart.png')
```

Write the data to accept the ungrouped data, then create an area chart for the data

```python
def daily_view(dataframe, company, date, color='purple'):
    df = dataframe.query(f'Company=="{company}" and Date=="{date}"').drop(columns = ['Company','Date']).melt()

    fig = df.plot(
        kind = 'area',
        figsize=(8,4),
        x='variable',
        y='value',
        color=f'{color}',
        title = f'Power Production: {company}/{date}',
        legend=False
    )
    fig.set_xlabel('Time of Day')
    fig.set_ylabel('TW/h Generated')
```

```python
daily_view(conc_t, '50Hertz', '2019-08-24', 'purple')
```

Renaming and dropping the columns to create so that they will have the axis renamed and .melt() to make it wide, instead of long

```python
a=hertz.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
b=amp.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
c=tennet.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
d=trans.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
```
